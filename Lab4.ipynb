```python
df_missing = df.copy()
df_missing.loc[0:5, 'units sold'] = np.nan
df_missing.isna().sum()
```

```python
print("Original shape: ",df.shape)
print("After removing some values: ",df_missing.shape)
```

```python
df_missing.head(10)
```

```python
df_imputed_median = df_missing.copy()
df_imputed_median['units sold'].fillna(df_imputed_median['units sold'].median())
df_imputed_median.isna().sum()
```

```python
df_imputed_median.head(10)
```

```python
The median was used to fill missing values because it is not
affected by extreme values or outliers, unlike the mean. This
ensures that the central tendency of the data is preserved
without being skewed by very large or very small numbers.
```

```python
plt.figure(figsize=(15, 4))  
plt.boxplot(df['units sold'], vert=False)  
plt.title('Box plot of units sold', fontsize=16)
plt.xlabel('Values', fontsize=14)
plt.show()
```

```python
Q1 = df['units sold'].quantile(0.25)
Q3 = df['units sold'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['units sold'] < lower) | (df['units sold'] > upper)]
outliers.head(15)
```

```python
df_no_outliers = df[(df['units sold'] >= lower) & (df['units sold'] <= upper)]
print("Original shape: ",df.shape)
print("After removing outliers: ",df_no_outliers.shape)
```

```python
lower_cap = df['units sold'].quantile(0.05)
upper_cap = df['units sold'].quantile(0.95)

df_capped = df.copy()
df_capped['units sold'] = df_capped['units sold'].clip(lower_cap, upper_cap)
```

```python
df[['units sold', 'gross sales']].head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_scaled = df[['units sold', 'gross sales']].copy()

df_scaled[['units sold', 'gross sales']] = scaler.fit_transform(df_scaled)

df_scaled.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_standardized = df[['units sold', 'gross sales']].copy()

df_standardized[['units sold', 'gross sales']] = scaler.fit_transform(df_standardized)

df_standardized.head()
```

```python
cols = ['units sold', 'gross sales']
numeric_df = df[cols]

corr = numeric_df.corr()

fig, ax = plt.subplots(figsize=(5,4))
cax = ax.matshow(corr, cmap='coolwarm', vmin=-1, vmax=1)  #  

fig.colorbar(cax)

ax.set_xticks(range(len(cols)))
ax.set_yticks(range(len(cols)))
ax.set_xticklabels(cols)
ax.set_yticklabels(cols)

for (i, j), val in np.ndenumerate(corr.values):
    ax.text(j, i, f"{val:.2f}", ha='center', va='center', color='black')

plt.title('Correlation Heatmap between units sales & gross sales')
plt.show()
```

```python
from sklearn.decomposition import PCA

X = df_standardized[['units sold', 'gross sales']]

pca = PCA(n_components=2)
principal_components = pca.fit_transform(X)

print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(principal_components[:,0], principal_components[:,1])
plt.title("PCA Projection")
plt.xlabel("units sold")
plt.ylabel("gross sales")
plt.show()
```
